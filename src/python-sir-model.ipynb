```python
import scipy.integrate
import numpy
import matplotlib.pyplot as plt
```

```python
def sir_model(y, t, beta, gamma):
    S, I, R = y
    
    dS_dt = -beta*S*I
    dI_dt = beta*S*I - gamma*I
    dR_dt = gamma*I
    
    return [dS_dt, dI_dt, dR_dt]
```

```python
# Total population, N.
N = 209.3*10**6 
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 190294, 1283
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.2, 1./10 
# A grid of time points (in days)

t = np.linspace(0, 160, 160)

# condições iniciais
S0 = 0.9
I0 = 0.1
R0 = 0.0
beta = 0.35
gamma = 0.1

# Time vector
t = numpy.linspace(0, 100, 10000)

# Result
solucao = scipy.integrate.odeint(sir_model, [S0, I0, R0], t, args=(beta, gamma))
solucao = numpy.array(solucao)
```

```python
plt.figure(figsize=[20,10])
plt.plot(t, solucao[:, 0], label="S(t)")
plt.plot(t, solucao[:, 1], label="S(t)")
plt.plot(t, solucao[:, 2], label="S(t)")
plt.grid()
plt.legend()
plt.xlabel("Tempo")
plt.ylabel("Proporção")
plt.title("SIR model")

plt.show()
```
